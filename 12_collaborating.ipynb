# Adding collaborating details

1. Open up Github, sign in and go to the *https://github.com/RSE-NI/eeecs_qub_collaboration* repo.  Click the green "use this template" button to recreate the repository and name it `eeecs_qub_collaboration`.  
2. In the `add_two_nums.py` file, the developer has left a TODO.  Turn this into an issue by going to issues, create new issue and pasting `add print statement` as the title and text.  You can then add a label, such as `enhancement` or `good first issue`. Save the issue and go back to the main repo.
3. Click Add File 



4. Create a new file called `CONTRIBUTIONS.md`  
5. In this, put details of how you want contributors to help.  A good example is https://github.com/Southampton-RSG-Training/shell-novice/blob/main/CONTRIBUTING.md.  
6. Remember to commit your changes.
