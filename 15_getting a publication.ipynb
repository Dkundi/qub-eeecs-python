# Software publication workflow  

1. Open up Github, sign in and go to the *https://github.com/RSE-NI/eeecs_qub_collaboration* repo.  If you haven't previously done the following, do it now: Click the green "use this template" button to recreate the repository and name it `eeecs_qub_collaboration`.  
2. Go to https://sandbox.zenodo.org/ and register for an account.  It's *very important* that the *sandbox* is used for training, not the main Zenodo site.
3. Click your profile name at the top right and select Github.  
4. Select connect and it will attempt to access your repositories (as long as you're signed in) with your permission.
5. When successful, it will redirect to sandbox.zenodo.com, with 3 steps to take.  
6. Flip the permission toggle on your eeecs_qub_collaboration repo to allow Zenodo to access it.  
7. Go back to Github and create a release using the link under Releases at the right hand side.  It is prudent to give it a proper title, tag (v1.0.0) and descriptive text.  
8. Go back to sandbox.zenodo.com and click the 'sync now' button.  
9. This should now show the repo details with a DOI for the exact version of the release.  As you're using the sandbox, hopefully, this is temporary.  



10. You can click the repo link and select the code for the DOI badge, then paste this into your README.  
11. Now add a `CITATION.cff` file to your repository, filling in the details:  

```
cff-version: 1.2.0
message: "If you use this software, please cite it as below."
authors:
  - family-names: 
    given-names: 
    orcid: 
title: ""
version: 
doi: 
date-released:
```

12. Commit the file and the repository should update to include "cite this" at the right hand side.  This will give you the bibtex and text version of the citation, allowing anyone to cite this software.
